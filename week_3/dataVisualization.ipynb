```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
import json
```

```python
df = pd.read_csv('../91App/91App_Datasets/Orders.csv', low_memory=False)
```

```python
df['new_DateId'] = pd.to_datetime(df['DateId'].astype(str), format='%Y%m%d')
df.head(5)
```

```python
data = df.groupby(by='new_DateId')['Quantity'].sum().to_frame()
data.reset_index(level=0, inplace=True)
data.head()
```

```python
import datetime
data['weekday'] = data['new_DateId'].apply(lambda x: x.weekday())
data.head()
```

```python
weekday_data = data.groupby(by='weekday')['Quantity'].sum().to_frame()
weekday_data.reset_index(level=0, inplace=True)
weekday_data.head()
```

```python
%matplotlib inline
plt.figure(figsize=(14,7))
plt.bar(weekday_data.weekday,weekday_data.Quantity, width=0.6, color="lightblue", zorder=2) 
plt.xlabel("Weekday", fontsize=18) 
plt.ylabel("Quantity", fontsize=18) 
plt.title("Sales Volume", fontsize=22)
weekday_labels = ["Sun","Mon","Tue","Wen","Thu","Fri","Sat"]
plt.xticks(weekday_data.weekday,weekday_labels, size=14)
plt.yticks(size=14)
plt.grid(True, zorder=1)
plt.show()
```

```python
movies_info = pd.read_json("movie.json")
```

```python
movies_info.head()
```

```python
movies_info.push.unique()
```

```python
movies_info.loc[movies_info["push"].str.match("爆"),"push"] = "100"
movies_info.loc[movies_info["push"].str.match("X1"),"push"] = "-10"
movies_info.loc[movies_info["push"].str.match("X2"),"push"] = "-20"
movies_info.loc[movies_info["push"].str.match("X3"),"push"] = "-30"
movies_info.loc[movies_info["push"].str.match("X4"),"push"] = "-40"
movies_info.loc[movies_info["push"].str.match("X5"),"push"] = "-50"
movies_info.loc[movies_info["push"].str.match("X6"),"push"] = "-60"
movies_info.loc[movies_info["push"].str.match("X7"),"push"] = "-70"
movies_info.loc[movies_info["push"].str.match("X8"),"push"] = "-80"
movies_info.loc[movies_info["push"].str.match("X9"),"push"] = "-90"
movies_info.loc[movies_info["push"].str.match("XX"),"push"] = "-100"
```

```python
movies_info['push'] = movies_info['push'].astype(str).astype(int)
movies_info.push.unique()
```

```python
movie_push = movies_info.groupby("date", as_index=False).sum()
movie_push = movie_push[~movie_push["date"].str.match("12/31")]
```

```python
movie_push
```

```python
%matplotlib inline
plt.figure(figsize=(14,7))
plt.title('Result Analysis')
plt.plot(movie_push.date, movie_push.push, color='green', label='training accuracy')
plt.legend()
plt.xlabel('dates')
plt.ylabel('push')
plt.legend(frameon=False)
plt.xticks(rotation='vertical')
plt.show()
```
